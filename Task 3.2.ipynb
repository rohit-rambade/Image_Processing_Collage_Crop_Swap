```python
# Import modules
import cv2
import matplotlib.pyplot as plt
```

```python
def show(p):
    cv2.imshow("phtoo", p)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
#image 1
img1 = cv2.imread("joker.jpg")
show(img1)
plt.imshow(img1)
img1.shape
```

```python
#image 2
img2 = cv2.imread("anonymous.jpg")
show(img2)
plt.imshow(img2)
img2.shape
```

```python
#Crop 1st Image
cropped_img1 = img1[50:650 ,650:1250]
show(cropped_img1)
```

```python
#crop 2nd Image
cropped_img2 = img2[50:650 ,650:1250]
show(cropped_img2)
```

```python
# Swapping the faces
temp = cropped_img1 + cropped_img2
img1[50:650 ,650:1250] = cropped_img2
show(img1)
```

```python
img2[50:650 ,650:1250] = temp-cropped_img2
show(img2)
```
