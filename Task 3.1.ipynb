```python
#Import packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def show(p):
    cv2.imshow("Image", p)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
#image 1
pic1 = mpimg.imread('omen.jpg')
plt.title('Orignal')
plt.imshow(pic1)
show(pic1)
```

```python
#Image2
pic2 = mpimg.imread('Chamber.jpg')
plt.title('Orignal')
plt.imshow(pic2)
show(pic2)
```

```python
#Now Combine the image
merge = np.hstack((pic1,pic2))
```

```python
plt.imshow(merge)
show(merge)
```
